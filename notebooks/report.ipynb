This notebook justifies the feature selection chosen in the Titanic Data Journey notebook  
To predict survival, the gender and the passenger class features were chosen

```python
import pandas as pd
import sys
import os
import matplotlib.pyplot as plt

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
import src.data as data
import src.visualize as visualize
```

```python
TRAIN_PATH = "../data/train.csv"
TEST_PATH  = "../data/test.csv"
FIG_DIRECTORY = "../figures/"
```

```python
train_df = data.load_data(TRAIN_PATH)
test_df = data.load_data(TEST_PATH)
all_df = pd.concat([train_df, test_df], axis=0)
all_df["set"] = "train"
all_df.loc[all_df.Survived.isna(), "set"] = "test"
```

```python
visualize.plot_count_pairs(all_df, "Sex", "Sex", hue="Survived")
visualize.save_recent_plot("sex_survived.png", FIG_DIRECTORY)
```

In this graph, we see that the proportion of female passengers who survived is much greater than the proportion of male passengers who survived.  
This suggests that gender is an important factor in predicting whether a passenger survived or not.

```python
visualize.plot_count_pairs(all_df, "Pclass", "Passenger Class", hue="Survived")
visualize.save_recent_plot("class_survived.png", FIG_DIRECTORY)
```

Pclass of 1 refers to first class (the best), while 3 refers to third class (the lowest).  
This graph shows that those from the best passenger class had a much higher survival chance than the lowest class, with over 50% of those from first class surviving, while a much lower proportion of the third class passengers survived.

```python
visualize.plot_distribution_pairs(data_df=all_df, feature="Age", title="Age", hue="Survived")
visualize.save_recent_plot("age_survived.png", FIG_DIRECTORY)
```

In this graph, we plot the distribution of age by survival status.
Across most age brackets, the number of people who survived is uniformly a 'little bit' less than the number of people who didn't survive. The exception is people under the age of 5, who were much more likely to have survived the incident.  

As a result, age as a continuous variable was not used in the prediction, but one may consider putting a 'young child' indicator to capture the observation that babies were more likely to survive.
